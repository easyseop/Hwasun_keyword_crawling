```python
def search(results):
    dic = defaultdict(list)
    for result in results.find_all('tr'):
        lis = result.find_all('td')


        dic['업무'].append(lis[0].text)
        dic['공고번호차수'].append(lis[1].text)
        dic['분류'].append(lis[2].text)
        dic['공고명'].append(lis[3].text)
        dic['공고기관명'].append(lis[4].text)
        dic['수요기관명'].append(lis[5].text)
        dic['계약방법'].append(lis[6].text)
        temp = lis[7].text.replace('(',' ').replace(')','')
        temp = temp.split()
        if len(temp)>2:

            dic['공고일시'].append(temp[0] +' '+ temp[1])
            dic['마감일시'].append(temp[2] +' '+ temp[3])
        else:
            dic['공고일시'].append(temp[0] +' '+ temp[1])
            dic['마감일시'].append('-')
    return dic
```

```python
import time
import requests
import pandas as pd
from fake_useragent import UserAgent
from bs4 import BeautifulSoup as bs
import urllib.request as ur
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from time import sleep
import warnings
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
warnings.filterwarnings("ignore")
from collections import defaultdict

basic_url = 'https://www.g2b.go.kr:8341/bp/bidPblancSearchList.do?areaLimit=&areaLimitNm=&bidInfo=&bidMethodCd=&bidNm=%B5%F0%C1%F6%C5%D0&cntcSysTyCode=&currMonthRadio=1&estimoney=&estimoneyCompare=U&fromBidDt=2023%2F01%2F15&industry=&industryCd=&instCl=1&instName=&intrlBidClCd=&maxPageViewNoByWshan=4&orderInstCd=&procmntReqNo=&recordCountPerPage=10&regYn=Y&searchClCd=search1&searchDtType=2&taskClCd=0&toBidDt=2023%2F02%2F14&currentPageNo='
page_num = 1
while True:
    url = basic_url + str(page_num)
    try:
        html = ur.urlopen(url)
        soup = bs(html.read(), "html.parser")
        results = soup.find('div','results').find('tbody')
        dic = search(results)
        pd.DataFrame(dic).to_csv(f'입찰공고목록_{page_num}.csv',encoding='utf-8-sig')
        page_num += 1
    except:
        print('HTTP Error 404: Not Found')
        break
    
```
